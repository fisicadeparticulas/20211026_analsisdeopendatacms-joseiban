# **Montecarlo para masa masa invariante del bosón $W^{\pm}$**

Usando la definición para la masa invariante, dada por:
\begin{equation*}
    M^2=2P_{T1}P_{T2}[\cosh{(\eta_1-\eta_2)}-\cos{(\phi_1-\phi_2)}]
\end{equation*}
Con nuestras definiciones para:
\begin{align*}
    P_T&=\sqrt{P_x^2+P_y^2} & \eta&=-\ln{\left[\tan{\left(\frac{\theta}{2}\right)}\right]} & \theta&=\arccos{\left(\frac{P_z}{\vec{P}}\right)} & \vec{P}&=\sqrt{P_T^2+P_z^2} &\phi&=\left[\frac{P_y}{|P_y|}\right]\arccos{\left(\frac{P_x}{P_T}\right)}
\end{align*}
Todas dadas en términos de las componentes cartesianas del momento.

## **Bibliotecas utilizadas**

```python
import pandas as pd #<--- GRACIAS A LA DOCTORA ISABEL PEDRAZA
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px #<--- GRACIAS A alancovamo
```

## **Masa de $W^+$**

### **Datos**

Archivo de datos emparejados $\mu^+$, $\nu_\mu$:

```python
#NOMBRAMIENTO DE DATOS:
data=np.genfromtxt("/content/drive/MyDrive/Masa Bosón W/mu+vm (1)_emp.csv", delimiter="," , names=["amu","pxam","pyam","pzam","Eam","vmu","pxv","pyv","pzv","Ev"]) 
#LECTURA DE DATOS:
eventos=pd.read_csv('/content/drive/MyDrive/Masa Bosón W/mu+vm (1)_emp.csv')
#IMPRESIÓN DE DATOS:
eventos.head(100000) #<--- NOS DA EL NÚMERO DE EVENTOS
```

### **Componentes de momento**

Componentes de $\mu^+$:

```python
plt.title("Px DE MU+")
plt.hist(data['pxam'])
np.nanmin(data['pxam']), np.nanmax(data['pxam'])
```

```python
plt.title("Py DE MU+")
plt.hist(data['pyam'])
np.nanmin(data['pyam']), np.nanmax(data['pyam'])
```

```python
plt.title("Pz DE MU+")
plt.hist(data['pzam'])
np.nanmin(data['pzam']), np.nanmax(data['pzam'])
```

### **Momentos**

El momento transverso para $\mu^+$:

```python
PTam=np.sqrt(data['pxam']**2+data['pyam']**2)
plt.title("MOMENTO TRANSVERSO DE MU+")
plt.hist(PTam)
np.nanmin(PTam), np.nanmax(PTam)
```

El momento total de $\mu^+$:

```python
Pam=np.sqrt(PTam**2+data['pzam']**2)
plt.title("MOMENTO TOTAL DE MU+")
plt.hist(Pam)
np.nanmin(Pam), np.nanmax(Pam)
```

El momento transverso para $\nu_\mu$:

```python
PTv=np.sqrt(data['pxv']**2+data['pyv']**2)
plt.title("MOMENTO TRANSVERSO DE VMU")
plt.hist(PTv)
np.nanmin(PTv), np.nanmax(PTv)
```

### **Pseudorapideces**

La pseudorapidez para $\mu^+$:

```python
Pam=np.sqrt(PTam**2+data['pzam']**2)
Thetaam=np.arccos(data['pzam']/Pam)
Etaam=-np.log(np.tan(Thetaam/2))
plt.title("ETA DE MU+")
plt.hist(Etaam)
np.nanmin(Etaam), np.nanmax(Etaam)
```

La pseudorapidez de $\nu_\mu$:

```python
Pv=np.sqrt(PTv**2+data['pzv']**2)
Thetav=np.arccos(data['pzv']/Pv)
Etav=-np.log(np.tan(Thetav/2))
plt.title("ETA DE VMU")
plt.hist(Etav)
np.nanmin(Etav), np.nanmax(Etav)
```

### **Ángulos respecto al eje $x>0$**

El ángulo $\phi$ para $\mu^+$:

```python
Phiam=(data['pyam']/np.abs(data['pyam']))*np.arccos(data['pxam']/PTam)
plt.title("ÁNGULO PHI DE MU+")
plt.hist(Phiam)
np.nanmin(Phiam), np.nanmax(Phiam)
```

Ángulo $\phi$ de $\nu_\mu$:

```python
Phiv=(data['pyv']/np.abs(data['pyv']))*np.arccos(data['pxv']/PTv)
plt.title("ÁNGULO PHI DE VMU")
plt.hist(Phiv)
np.nanmin(Phiv), np.nanmax(Phiv)
```

### **Ángulo entre las trazas de las partículas**

El ángulo $\Delta\phi$:

```python
Dphi=((np.abs(data['pyam'])/data['pyam'])*Phiam)+((np.abs(data['pyv'])/data['pyv'])*Phiv)
plt.title("ÁNGULO DELTA PHI")
plt.hist(Dphi, bins=[3.125,3.13,3.135,3.14,3.145,3.15,3.155,3.16])
np.nanmin(Dphi), np.nanmax(Dphi)
```

### **Masas**

Masa invariante:

```python
MI=np.sqrt(2*PTam*PTv*(np.cosh(Etaam-Etav)-np.cos(Phiam-Phiv)))
plt.title("MASA INVARIANTE DE W+")
plt.hist(MI)
np.nanmin(MI), np.nanmax(MI)
```

```python
MI1=px.histogram(MI,MI)
MI1.show()
```

Masa transversa: $M_T=\sqrt{2P_{T1}P_{T2}(1-\cos{\Delta\phi})}$

```python
MT=np.sqrt(2*PTam*PTv*(1-np.cos(Dphi)))
plt.title("MASA INVARIANTE DE W+")
plt.hist(MT)
np.nanmin(MT), np.nanmax(MT)
```

```python
MT1=px.histogram(MT,MT)
MT1.show()
np.nanmin(MT), np.nanmax(MT)
```

```python
PTm=np.sqrt(data['pxm']**2+data['pym']**2)
PTav=np.sqrt(data['pxav']**2+data['pyav']**2)
Dphi1=np.arccos(data['pxm']/PTm)+np.arccos(data['pxav']/PTav)
MTm=np.sqrt(2*PTm*PTav*(1-np.cos(Dphi1)))
plt.title("MASA TRANSVERSA DE W-")
plt.hist(MTm)
np.nanmin(MTm), np.nanmax(MTm)
```

```python
MTm1=px.histogram(MTm,MTm)
MTm1.show()
np.nanmin(MTm), np.nanmax(MTm)
```
